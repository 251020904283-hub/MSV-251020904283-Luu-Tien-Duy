```
# Bài 1
n = int(input("Nhập vào số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
  if i % 2  ==0:
    tong = tong + i
print("Tổng các số chẵn nhỏ hơn hoặc bằng" , n , "là" , tong)
```

```
# Bài 2
s = input("Nhập vào một chuỗi: ")
a = s.split()
b= len(a)
c = a[0]
for d in a:
    if len(d) > len(c):
        c = d
print("Số từ trong chuỗi:", b)
print("Từ dài nhất:", c)
```

```
# Bài 3
a = list(map(int, input("Nhập các số: ").split()))
b = []
for x in a:
    if x not in b:
        b.append(x)
b.sort()
c = b[-2]

print("Danh sách sau khi sắp xếp:", b)
print("Số lớn thứ hai:", c)
```

```
# Bài 4
x = list(map(float, input("Nhập điểm: ").split()))
y = []
for d in x:
    if 0 <= d <= 10:
        y.append(d)
tong = 0
for d in y:
    tong += d
z = tong / len(y)
dat = 0
for d in y:
    if d >= 5:
        dat += 1

print("Điểm trung bình:", z)
print("Số sinh viên đạt:", dat)
```
